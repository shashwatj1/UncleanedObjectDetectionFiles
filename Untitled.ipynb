```python
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

class CocoSegmentation:
    def __init__(self, json_file, image_folder):
        self.json_file = json_file
        self.image_folder = image_folder
        self.coco_data = self.load_coco_annotations()
        self.class_names = self.load_class_names()

    def load_coco_annotations(self):
        with open(self.json_file, 'r') as f:
            coco_data = json.load(f)
        return coco_data

    def load_class_names(self):
        class_names = {}
        for category in self.coco_data['categories']:
            class_names[category['id']] = category['name']
        return class_names

    def get_image_by_id(self, image_id):
        for image in self.coco_data['images']:
            if image['id'] == image_id:
                return image
        return None

    def get_annotations_by_image_id(self, image_id):
        annotations = []
        for annotation in self.coco_data['annotations']:
            if annotation['image_id'] == image_id:
                annotations.append(annotation)
        return annotations

    def display_image_with_polygons(self, image_path, annotations):
        img = Image.open(image_path)
        plt.imshow(img)

        ax = plt.gca()

        for annotation in annotations:
            # Draw segmentation polygons
            if 'segmentation' in annotation and isinstance(annotation['segmentation'], list):
                for seg in annotation['segmentation']:
                    polygon = np.array(seg).reshape(-1, 2)

                    # Create a patch for the polygon with a light transparent fill
                    patch = patches.Polygon(polygon, linewidth=2, edgecolor='r', facecolor='r', alpha=0.3)
                    ax.add_patch(patch)

                    # Create a patch for the outline
                    outline_patch = patches.Polygon(polygon, linewidth=2, edgecolor='r', facecolor='none')
                    ax.add_patch(outline_patch)

            # Draw bounding box
            x, y, w, h = annotation['bbox']
            bbox_patch = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='b', facecolor='none')
            ax.add_patch(bbox_patch)

            # Add class name text
            class_id = annotation['category_id']
            class_name = self.class_names[class_id]
            ax.text(x, y, class_name, fontsize=10, color='b', bbox=dict(facecolor='white', alpha=0.5))

        plt.axis('off')
        plt.show()

    def display_segmentation_only(self, image_path, annotations):
        img = Image.open(image_path)
        plt.imshow(img)

        ax = plt.gca()

        for annotation in annotations:
            if 'segmentation' in annotation and isinstance(annotation['segmentation'], list):
                for seg in annotation['segmentation']:
                    polygon = np.array(seg).reshape(-1, 2)
                    patch = patches.Polygon(polygon, linewidth=2, edgecolor='r', facecolor='r', alpha=0.3)
                    ax.add_patch(patch)

        plt.axis('off')
        plt.show()

    def display_bboxes_only(self, image_path, annotations):
        img = Image.open(image_path)
        plt.imshow(img)

        ax = plt.gca()

        for annotation in annotations:
            x, y, w, h = annotation['bbox']
            bbox_patch = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='b', facecolor='none')
            ax.add_patch(bbox_patch)

            class_id = annotation['category_id']
            class_name = self.class_names[class_id]
            ax.text(x, y, class_name, fontsize=10, color='b', bbox=dict(facecolor='white', alpha=0.5))

        plt.axis('off')
        plt.show()

    def extract_and_display_polygons(self, image_id):
        image_info = self.get_image_by_id(image_id)
        if not image_info:
            print(f"Image ID {image_id} not found in annotations.")
            return

        image_path = f"{self.image_folder}/{image_info['file_name']}"
        annotations = self.get_annotations_by_image_id(image_id)

        self.display_image_with_polygons(image_path, annotations)

    def extract_and_display_segmentation_only(self, image_id):
        image_info = self.get_image_by_id(image_id)
        if not image_info:
            print(f"Image ID {image_id} not found in annotations.")
            return

        image_path = f"{self.image_folder}/{image_info['file_name']}"
        annotations = self.get_annotations_by_image_id(image_id)

        self.display_segmentation_only(image_path, annotations)

    def extract_and_display_bboxes_only(self, image_id):
        image_info = self.get_image_by_id(image_id)
        if not image_info:
            print(f"Image ID {image_id} not found in annotations.")
            return

        image_path = f"{self.image_folder}/{image_info['file_name']}"
        annotations = self.get_annotations_by_image_id(image_id)

        self.display_bboxes_only(image_path, annotations)
```

```python
coco_json_file = 'Dataset/annotations_trainval2017/annotations/instances_val2017.json'
image_folder = 'Dataset/val2017/val2017/'
coco_segmentation = CocoSegmentation(coco_json_file, image_folder)
image_id = 16228  
```

```python
coco_segmentation.extract_and_display_segmentation_only(image_id)
```

```python
coco_segmentation.extract_and_display_bboxes_only(image_id)
```
